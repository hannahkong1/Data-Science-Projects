## 4. Summarizing Data

Research Question: What is the relationship between US Sales(in millions) of a video game and its publisher?

```python
df_publisher = pd.DataFrame(df[['US Sales (millions)', 'Publisher']].groupby('Publisher').mean())
```

```python
df_publisher.sort_values(by = 'US Sales (millions)', ascending = False).head(20)
```

```python
##visualization
import seaborn as sns
sns.boxplot(x = 'US Sales (millions)', data = df_publisher)
plt.show()
```

```python
##summary statistics
df_publisher['US Sales (millions)'].describe()
```

Based off all of the analysis I did above: group by to find the medians of US Sales for each publisher, a box plot showing the distribution of US Sales across the dataset, and the summary statistics of US Sales for all the video games in the dataset, I can see that larger US Sales might be correlated with big publishing names. In df_publisher, I grouped the US Sales in millions by each publisher that is included in the dataset. From the top 20 publishers with the highest US Sales, I can clearly see that 10 of these 20 (50%) are all Nintendo. Using outside knowledge, I know that Nintendo is a very popular video game publisher. Also in the box plot, most of the points that seem to stray away from the maximum value are from the games that are published by Nintendo. So from all this information, I am concluding that with a bigger publisher name, there is a chance that that specific game will get higher US Sales and have a larger change of being a blockbuster video game.
